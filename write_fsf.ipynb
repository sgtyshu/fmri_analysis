```python
import re

# 读取fsf文件
file_path = '/nfs/h1/userhome/liyifan/workingdir/Raven-fmri/fsf/group_anova/level2_group_anova.fsf'
with open(file_path, 'r') as file:
    fsf_content = file.readlines()

# 查找匹配的行并替换
modified_fsf_content = ""
for i in range(len(fsf_content)):
    line = fsf_content[i]
    if line.startswith("#") and "input" in line:
        match = re.search(r"input\s+(\d+)", line)
        if match:
            input_number = int(match.group(1))
            if input_number % 10 == 1 and i + 1 < len(fsf_content):
                next_line = fsf_content[i + 1]
                next_line_modified = re.sub(r"(\d+)\s*$", "1", next_line)
                modified_fsf_content += next_line_modified
                continue
    modified_fsf_content += line

# 将修改后的内容写回fsf文件
modified_file_path = '/nfs/h1/userhome/liyifan/workingdir/Raven-fmri/fsf/group_anova/modified.fsf'
with open(modified_file_path, 'w') as file:
    file.write(modified_fsf_content)
```
